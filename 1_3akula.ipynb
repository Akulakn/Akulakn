```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sp
from sympy import init_printing
init_printing()

from sympy import symbols, Eq, Function



from IPython.display import display
```

```python
M=4
m=1
T=10
L=1
g=9.8
#f=np.sin(5*t)
#x(0)=0
#Θ(0)=Pi/15
#Θ(T)=0
#x(T)=0

def f(t,y):
    return np.vstack((y[2], y[3], 2*((np.sin(5*t)-(y[3]*(M+m)))/(M*L)), g*y[1]-((7/12)*L*y[2])))

# граничные условия
def bc(ya, yb):
    return np.array([ya[0] + 0.15, ya[2], yb[0] + 3.14/15,  yb[2]])

# Сетка значений t на отрезке [0;T]
t = np.linspace(0, T, 100)

# Начальное приближение значений функции y в узлах сетки
y = np.zeros((4, t.size))

res = integrate.solve_bvp(f, bc, t, y)

t_plot = np.linspace(0, T, 100)
y_plot = res.sol(t_plot)[0]
plt.plot(t_plot, y_plot, 'r')
plt.xlabel("t")
plt.ylabel("x")
plt.grid()
plt.show()

t_plot = np.linspace(0, T, 100)
y_plot = res.sol(t_plot)[1]
plt.plot(t_plot, y_plot, 'y')
plt.xlabel("t")
plt.ylabel("Θ")
plt.grid()
plt.show()

t_plot = np.linspace(0, T, 100)
y_plot = res.sol(t_plot)[2]
plt.plot(t_plot, y_plot, 'b')
plt.xlabel("t")
plt.ylabel("v")
plt.grid()
plt.show()

t_plot = np.linspace(0, T, 100)
y_plot = res.sol(t_plot)[3]
plt.plot(t_plot, y_plot,'k')
plt.xlabel("t")
plt.ylabel("w")
plt.grid()
plt.show()
```
