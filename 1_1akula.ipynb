```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import integrate
import sympy as sp
from sympy import init_printing
init_printing()

from sympy import symbols, Eq, Function



from IPython.display import display
```

```python
def f(t,r):
    x, y, u, v = r
    fx = u
    fy = v
    fu = ( -1*x + 2*y) / 1
    fv = -x 
    return fx, fy, fu, fv

fig, ax = plt.subplots(figsize=(20, 10))
sol = integrate.solve_ivp(f, (0, 40), (1, 2, 3, 4), t_eval = np.linspace(0,30,200)) 
x, y, u, v = sol.y
t = sol.t

ax.plot(t, x)
ax.set_xlabel(r"$t$", fontsize=25)
ax.set_ylabel(r"$x 1,2$", fontsize=25)
ax.plot(t, x, label=r"x1")

ax.plot(t, y)
ax.plot(t, y, label=r"x2")
ax.legend(fontsize=25)

ax.grid()

fig, ax = plt.subplots(figsize=(20, 10))
ax.plot(t, u)
ax.set_xlabel(r"$t$", fontsize=25)
ax.set_ylabel(r"$v 1,2$", fontsize=25)
ax.plot(t, u, label=r"v1")

ax.plot(t, v)
ax.plot(t, v, label=r"v2")
ax.legend(fontsize=25)

ax.grid()
```
