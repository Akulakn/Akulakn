```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
#правая часть системы
def f(t,y):
    return np.vstack((y[1], -y[1]/t))

#граничные условия
def bc(ya, yb):
    return np.array([ya[0], yb[0]-1])

#отрезок
T= np.exp(1)
t = np.linspace(0.00000001,T, 100)


y = np.zeros((2, t.size))

res = integrate.solve_bvp(f, bc, t, y)

t_plot = np.linspace(0.00000001,T, 100)
y_plot = res.sol(t_plot)[0]
plt.plot(t_plot, y_plot)
plt.xlabel("t")
plt.ylabel("y")
plt.grid()
plt.show()
```
